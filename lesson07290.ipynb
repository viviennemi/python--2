```python
#標準寫法--先安裝繪圖軟體
import matplotlib.pyplot as plt

##畫圖前1.2都一定要弄出來
# 1.製作figure--圖表
fig = plt.figure()

# 2.在1.的figure上製作一個axes
#subplot(1, 1, 1)-->X軸一個 Y軸一個 畫哪一個圖
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中--plot是線圖
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

figure-畫布 
Axes-圖表
X-axis X軸
Y-axis Y軸
ticks label 刻度標籤
ticks顯示刻度
annotation標註

```python
fig = plt.figure()
ax_1 = fig.add_subplot(1,3,1)
ax_1.set_title("axes1")

ax_2 = fig.add_subplot(1,3,2)
ax_2.set_title("axes2")

ax_3 = fig.add_subplot(1,3,3)
ax_3.set_title("axes3")
```

```python
student1 = {'Chinese':80, 'English': 63, 'Math':93, 'History':85, 'Discovery':78}
student2 ={'Chinese':85,'English':83,'Math':88,'History':80,'Discovery':88}
student3 ={'Chinese':80,'English':83,'Math':78,'History':80,'Discovery':84}
student4 ={'Chinese':70,'English':73,'Math':78,'History':70,'Discovery':74}
subjects = list(student1.keys())
value1 = list(student1.values())
value2 = list(student2.values())
value3 = list(student3.values())
value4 = list(student4.values())
##顏色前面要加#一樣的字母6個就為灰色底
figure = plt.figure(figsize=(8.0,5.0),dpi=250.0,facecolor="#aaaaaa")
ax = figure.add_subplot()
ax.plot(subjects, value1,'go--', linewidth=1, markersize=5)
ax.plot(subjects,value2,color='brown', marker='x', linestyle='--',
     linewidth=2, markersize=5)
#The optional parameter fmt is a convenient way for defining basic formatting like color, marker and linestyle.
#顏色。標記跟線條的部分可以不需要給定義直接寫出來讓程式判斷，其他的部分都需要寫出名稱作為定義。
#It's a shortcut string notation described in the Notes section below.
#顏色的定義https://matplotlib.org/stable/gallery/color/named_colors.html
ax.plot(subjects, value3,'m--^')
ax.plot(subjects, value4,'y-.o',linewidth=5,markersize=10)
plt.show()
```

```python
student1 = {'Chinese':80, 'English': 63, 'Math':93, 'History':85, 'Discovery':78}
subjects = list(student1.keys())
value1 = list(student1.values())
figure = plt.figure(figsize=(8.0,5.0),dpi=250.0,facecolor="#aaaaaa")
ax = figure.add_subplot()
ax.plot(subjects, value1,'go--', linewidth=1, markersize=5)
print(ax.axis())
ax.annotate("student1",(3.5,80))
plt.show()
print(ax.get_xticks())
ax.get_yticks()
```
